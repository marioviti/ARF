# Support Vectors on different Kernels function.

As function kernels evolve, the number of support vectors increases.
From the SVM primal problem, we know that especially when having hard margins, the alphas are set to satisfy the constraint that $\sum_{i \in \textbf{D}}\textbf{y}^{i}(\textbf{wx}^{i}+b)\geq 1$, greater $\alpha$ and therefore support vectors are associated  with examples on or within the margins, in order to approximate a linear margin less points are needed than a polinomial or a radial function. 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

X,Y = arftools.gen_arti()

def plot_SVM(X,Y,kernels = ['linear'], C=0):
    fignum = 1
    #plt.suptitle("SVM on Separated normal distributed data with different kernels")
    for kernel in kernels:
        if C == 0:
            clf = svm.SVC(kernel=kernel)
        else:
            clf = svm.SVC(kernel=kernel, C=C)
        clf.fit(X, Y)
        plt.figure(fignum, figsize=(10, 6))
        plt.clf()
        svplot = plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=50, facecolors='red', marker='o', zorder=3, )
        plt.axis('tight')
        plt.figure(fignum, figsize=(4, 3))
        arftools.fig_plot_frontiere(plt,X,clf.decision_function)
        arftools.fig_plot_data(plt,X,Y)
        plt.legend([svplot],['support vectors'],scatterpoints=1)
        fignum = fignum + 1
    plt.show()
    
plot_SVM(X,Y,kernels = ['linear', 'poly', 'rbf'])
```

## Multiclass classification.

We'll test the SVM kernels on the USPS dataset of handwritten digits and we'll plot the ROC curve for the test and relatives AUC to estimate accuracy and precision.

```python
import USPS_handler as USPS
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

fn_usps_train = 'USPS_train.txt'
fn_usps_test = 'USPS_test.txt'

trainx,trainy = USPS.load_usps(fn_usps_train)
testx,testy = USPS.load_usps(fn_usps_test)

kernels = ['linear', 'poly', 'rbf']
clflin = svm.SVC(probability=True, kernel='linear')
train_scores_lin, test_scores_lin, roc_curves_lin =  USPS.usps_1vsMulti_class_train_and_test(trainx,trainy,testx,testy,clflin)

clfpoly = svm.SVC(probability=True, kernel='poly')
train_scores_poly, test_scores_poly, roc_curves_poly =  USPS.usps_1vsMulti_class_train_and_test(trainx,trainy,testx,testy,clfpoly)

clfrbf = svm.SVC(probability=True, kernel='rbf')
train_scores_clfrbf, test_scores_clfrbf, roc_curves_clfrbf =  USPS.usps_1vsMulti_class_train_and_test(trainx,trainy,testx,testy,clfrbf)

```

```python
def plot_ROCS_multiclass(roc_curves_values,title='Receiver operating characteristic example'):
    labels = ['digit %d vs all AUC %f' % (i,roc_curves_values[i][2]) for i in range(len(roc_curves_values))]
    colors = ['blue','red','orange','green','yellow','darkorange','pink','purple','cyan','magenta']
    plot_ROCS(roc_curves_values,colors,labels,title=title)

def plot_ROCS(roc_curves_values,colors,labels,title='Receiver operating characteristic example'):
    plt.figure(0, figsize=(10, 6))
    plots = []
    for i in range(len(roc_curves_values)):
        fpr,tpr,roc_auc = roc_curves_values[i]
        plot, = plt.plot(fpr, tpr, color=colors[i], label=labels[i])
        plots += [plot]
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.legend(handles=plots)
    plt.grid(True)
    plt.title(title)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    
plot_ROCS_multiclass(roc_curves_lin, title='Receiver operating characteristic SVM linear')
plt.show()

plot_ROCS_multiclass(roc_curves_poly, title='Receiver operating characteristic SVM poly')
plt.show()

plot_ROCS_multiclass(roc_curves_clfrbf, title='Receiver operating characteristic SVM rbf')
plt.show()
    
```
